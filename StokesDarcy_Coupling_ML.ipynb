# Prediction of the Macroscopic Permeability of doubly porous media with non-interconnected macropores

```python
!pip install gradio
from sklearn.ensemble import GradientBoostingRegressor
!pip install scikit-learn --upgrade
import pandas as pd
import numpy  as np
import gradio as gr
import pickle
from google.colab import drive
drive.mount('/content/drive')
%cd /content/drive/MyDrive/
```

```python
drive.mount('/content/drive')
%cd /content/drive/MyDrive/
def make_prediction(X1,X2,X3,X4,X5):
    with open("StokesDarcyML.pkl", "rb") as f:
        model  = pickle.load(f)
        Test_Empi1=pd.DataFrame(np.array([[X1,X2,X3,X4,X5]]),
                                columns=['a','b','m','n','knano'])
        preds = model.predict(Test_Empi1)
    return preds

#Create the input component for Gradio
x1_input = gr.Slider(value=0.05, minimum=0.05, maximum=0.45, step=0.01,label = "a")
x2_input = gr.Slider(value=0.05, minimum=0.05, maximum=0.45, step=0.01,label = "b")
x3_input = gr.inputs.Dropdown(choices=[0.03, 1, 2, 50],label = "m")
x4_input = gr.inputs.Dropdown(choices=[0.03, 1, 2, 50],label = "n")
x5_input = gr.Slider(value=1e-6, minimum=1e-6,maximum=1e-4, step=1E-6,label = "Knano")

#Calculate the output
output = gr.Textbox(label = "Predicted macroscopic permeability")
app = gr.Interface(fn = make_prediction, inputs=[x1_input,x2_input,x3_input,
                                                 x4_input,x5_input], outputs=output)
app.launch(share=True)
```
